# Coding Logistic Regression for OR Gate

Importing the packages

```python
import numpy as np
import random 
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import warnings
import time
from mpl_toolkits import mplot3d
import math
```

Define Sigmoid Function

```python
def sigmoid(p):
    return 1 / (1 + math.exp(-p))
```

Ignoring Warning

```python
warnings.filterwarnings("ignore")
```

Define OR Gate

```python
x1= np.array([0,0,1,1]);
x2=np.array([0,1,0,1])
y= np.array([0,0,0,1]);
yh=np.array([0.1,0.22,0.33,0.4]);
```

Initialize Weight

```python
w2, w1, w0 = -5,-5,10;
```

Create the first set of output

```python

for i in range(0, 4):
    yh[i] = sigmoid(w0 + w1*x1[i] + w2 *x2[i] )
```

Cost Function

```python

j=[]
j.append(0);
# Binary Cross Entropy Cost Function
j.append(-1* sum(y * np.log(yh)+ (1-y) * np.log((1-yh))))
```

Define iterations

```python
itr = [];
itr.append(0);
itr.append(1)
```

Define learning rate, error threshold, coefficient for momentum for mommentum based gradient descent.

```python
alp=1;
ep= 0.001;
err, iter = 30 , 1;
r= 0.01
m=4; 
```

```python
while (err > ep and iter < 150):
    
    # Calculate Gradient based upon the output 
    
    DJ0= yh-y;
    DJ1 =(yh-y)* x1;
    DJ2 =(yh-y)* x2;
   
    # This is normal gradient based descent
    '''
    dw0=-alp* sum(DJ0) / m;
    dw1 = -alp* sum(DJ1) / m;
    dw2 = -alp* sum(DJ2) / m;
    '''
    # now I am introducing momentum based gradient descent. 25% less iterations but les accuracy fir weightages
    
    dw0=-(w0 * r +(alp* sum(DJ0) / m));
    dw1=-(w1 * r +(alp* sum(DJ1) / m));
    dw2=-(w2 * r +(alp* sum(DJ2) / m));
    
    
    # Update the weights
    w0= w0 + dw0;
    w1 = w1 + dw1;
    w2 = w2 + dw2;

    for i in range(0, 4):
        yh[i] = sigmoid(w0 + w1*x1[i] + w2 *x2[i] )
    
    iter = iter +1;
    itr.append(iter)
    
    # Calculate Cost Function
    
    j.append(-1* sum(y * np.log (yh)+ (1-y) * np.log (1-yh)))
    
    # j.append (0.5/m *  sum ((y-w1*x1-w0)**2) );
    
    err = abs(j[iter] - j[iter-1] );
    
    plt.subplot(211);
    plt.plot(itr,j,'o')
    plt.xlabel('iterations')
    plt.ylabel('cost')
    plt.show() 
    
```
